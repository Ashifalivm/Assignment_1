```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the dataset from Excel
data = pd.read_excel('C:/Users/Ashif/Downloads/iris.xls')
```

```python
# Display the columns in the dataset
print("Columns in the dataset:")
print(data.columns)
```

```python
# Calculate the mean of each column
mean_values = data.mean()
print("\nMean of each column:")
print(mean_values)
```

```python
# Check for null values
null_values = data.isnull().sum()
print("\nNull values in the dataset:")
print(null_values)
```

```python
# Visualization 1: Histogram of Sepal Length (SL)
plt.figure(figsize=(8, 6))
plt.hist(data['SL'], bins=10, color='skyblue')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Length')
plt.show()
```

```python
# Visualization 2: Scatter plot of Sepal Width (SW) vs. Petal Length (PL)
plt.figure(figsize=(8, 6))
plt.scatter(data['SW'], data['PL'], c='green', alpha=0.7)
plt.xlabel('Sepal Width')
plt.ylabel('Petal Length')
plt.title('Scatter Plot of Sepal Width vs. Petal Length')
plt.show()
```

```python
# Visualization 3: Box plot of Petal Width (PW) by Class
plt.figure(figsize=(8, 6))
plt.boxplot([data[data['Classification'] == 'Iris-setosa']['PW'],
             data[data['Classification'] == 'Iris-versicolor']['PW'],
             data[data['Classification'] == 'Iris-virginica']['PW']],
            labels=['Setosa', 'Versicolor', 'Virginica'])
plt.xlabel('Class')
plt.ylabel('Petal Width')
plt.title('Box Plot of Petal Width by Class')
plt.show()
```
